```python
import fitz
```

```python
import re
def extract_raw_text(pdf_path):
    # Open the PDF
    doc = fitz.open(pdf_path)
    
    # Initialize variables to store invoice details
    invoice_date = ""
    shop_name = ""
    total_amount = ""
    raw_text=""
    
    # Loop through each page of the PDF
    for page in doc:
        # Extract text from the page
        text = page.get_text()
        
        # Search for invoice details using regular expressions
        invoice_date_match = re.search(r"Date:\s*(.*)", text)
        shop_name_match = re.search(r"Shop Name:\s*(.*)", text)
        total_amount_match = re.search(r"Total:\s*(.*)", text)
        
        # Extract the matched groups from the regular expressions
        if invoice_date_match:
            invoice_date = invoice_date_match.group(1).strip()
        if shop_name_match:
            shop_name = shop_name_match.group(1).strip()
        if total_amount_match:
            total_amount = total_amount_match.group(1).strip()
        
        # Append the extracted text to the raw text
        raw_text += text
    
    # Close the PDF
    doc.close()
    
    # Return the invoice details
    return raw_text
```

```python
import re

def extract_fields(raw_text):
    # Define regular expressions to match each field
    date_pattern = r"\d{1,2}/\d{1,2}/\d{2,4}"
    invoice_number_pattern = r"Invoice\s*Number[:;]?\s*(\d+)"
    invoice_generator_pattern = r"Host[:;]?\s*(\w+)"
    name_pattern = r"Order\s*Total\n([A-Za-z\s]+)\n"
    due_date_pattern = r"Due\s*Date[:;]?\s*(\d{1,2}/\d{1,2}/\d{2,4})"
    items_pattern = r"\d+\.\d+\s+(.+)\n"

    # Extract fields using regular expressions
    date = re.search(date_pattern, raw_text).group() if re.search(date_pattern, raw_text) else None
    invoice_number = re.search(invoice_number_pattern, raw_text).group(1) if re.search(invoice_number_pattern, raw_text) else None
    invoice_generator = re.search(invoice_generator_pattern, raw_text).group(1) if re.search(invoice_generator_pattern, raw_text) else None
    name = re.search(name_pattern, raw_text).group(1) if re.search(name_pattern, raw_text) else None
    due_date = re.search(due_date_pattern, raw_text).group(1) if re.search(due_date_pattern, raw_text) else None
    items = re.findall(items_pattern, raw_text)

    return {
        "Date": date,
        "Invoice Number": invoice_number,
        "Invoice Generator": invoice_generator,
        "Name": name,
        "Due Date": due_date,
        "Items": items
    }

# Sample raw text
raw_text = """
Cowgirl Creamery
415-362-9354
The Ferry Plaza
One Embarcadero
T٥ NY
04/08/2019
Host: Gabino
11 :55
AM
TONY
40060
9.75
Far nouse GC
Hni i^ Made Lemonade
2.95
12.70
Subtotal
1
Tax: ;
Order Total
13.78
13.78
MC
Auth: NV97AY
Thank you for shopping at
Cowgirl Creamery!
Follow us on Instagram and Facebook
for the latest cheese news!
--- Check Closed ---
"""

```

```python
# Extract fields
fields = extract_fields(text)

# Print extracted fields
for key, value in fields.items():
    print(f"{key}: {value}")
```
